# List of movies & casting

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.228.0 Safari/537.36'
}
```

```python
url='https://www.imdb.com/list/ls058325506/'
```

```python
response=requests.get(url,headers=headers)
```

```python
soup=bs(response.content,'html.parser')
```

```python
soup
```

```python
name = h3, ipc-title__text
cast = span, sc-74bf520e-5 eesgaX-------a, ipc-link ipc-link--base dli-cast-item
```

```python
movie_name = soup.find_all('h3',class_='ipc-title__text')
```

```python
movie_name
```

```python
movies_list=[]
for i in movie_name:
    movies_list.append(i.text)
    
```

```python
movies_list
```

```python
for i in range(2):
    movies_list.pop()
print(movies_list)
```

```python
cast_names=[]
for i in soup.find_all('span',class_='sc-74bf520e-5 eesgaX'):
    cast_names.append(i.text)
print(cast_names)
```

```python
table=soup2.find('table',class_='wikitable')
```
